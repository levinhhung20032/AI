#11_matplotlib/company-sale

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
url="https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv"
data=pd.read_csv(url)
data
```

##Ex1

```python
plt.title("Company profit per month")
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.xticks([i for i in range(1,13)])
plt.yticks([i*100000 for i in range(1,6)])
plt.ylim(100000, 500000)
plt.plot(data['month_number'],data['total_profit'], color="b")
```

##Ex2

```python
plt.title("Company Sales data of last year")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks([i for i in range(1,13)])
plt.yticks([i*100000 for i in range(1,6)])
plt.ylim(100000, 500000)
plt.plot(data['month_number'],data['total_profit'],label="Profit data of last year",linewidth=3, color="r", linestyle = '--', marker = 'o', ms=8, mfc='k')
plt.legend(loc="lower right")
plt.show()
```

##Ex3

```python
plt.title("Sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.xticks([i for i in range(1,13)])
plt.yticks([i*1000 for i in [1,2,4,6,8,10,12,15,18]])
plt.ylim(0,18000)
plt.plot(data['month_number'],data['facecream'],label="Face cream Sales Data",linewidth=3, color="blue",marker = 'o')
plt.plot(data['month_number'],data['facewash'],label="Face Wash Sales Data",linewidth=3, color="orange",marker = 'o')
plt.plot(data['month_number'],data['toothpaste'],label="ToothPaste Sales Data",linewidth=3, color="green",marker = 'o')
plt.plot(data['month_number'],data['bathingsoap'],label="ToothPaste Sales Data",linewidth=3, color="red",marker = 'o')
plt.plot(data['month_number'],data['shampoo'],label="ToothPaste Sales Data",linewidth=3, color="purple",marker = 'o')
plt.plot(data['month_number'],data['moisturizer'],label="ToothPaste Sales Data",linewidth=3, color="brown",marker = 'o')
plt.legend(loc="upper left")
plt.show()
```

##Ex4

```python
plt.title("Tooth paste Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Number of units Sold")
plt.xticks([i for i in range(1,13)])
plt.scatter(data['month_number'],data['toothpaste'],label="Tooth paste Sales Data",linewidth=3, color="blue",marker = 'o')
plt.grid(color = 'gray', linestyle = '--')
plt.legend(loc="upper left")
plt.show()
```

##Ex5

```python
plt.title("Facewash and facecream sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.xticks([i for i in range(1,13)])
plt.bar(data['month_number']-0.1,data['facecream'],label="Face Cream sales data", width=0.2, color="blue")
plt.bar(data['month_number']+0.1,data['facewash'],label="Face Wash sales data", width=0.2, color="orange")
plt.grid(color = 'gray', linestyle = '--')
plt.legend(loc="upper left")
plt.show()
```

##Ex6

```python
plt.title("bathingsoap sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.xticks([i for i in range(1,13)])
plt.bar(data['month_number'],data['bathingsoap'], width=0.7, color="blue")
plt.grid(color = 'gray', linestyle = '--')
plt.show()
```

##Ex7

```python
plt.title("Profit data")
plt.xlabel("profit range in dollar")
plt.ylabel("Actual Profit in dollar")
plt.xticks([150000, 175000, 200000, 225000, 250000, 300000, 350000])
plt.hist(data['total_profit'], bins=[150000, 175000, 200000, 225000, 250000, 300000, 350000],label="Profit data", color="blue")
plt.legend(loc="upper left")
plt.show()
```

##Ex8

```python
plt.title("Sales data")
pie=data.drop(['month_number', 'total_units', 'total_profit'], axis=1)
plt.pie(pie.sum(axis=0), labels=['FaceCream','FaceWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer'], autopct='%1.2f%%')
plt.legend(loc="lower right")
plt.show()
```

##Ex9

```python
plt.subplot(2,1,1)
plt.title("Sales data of a Bathingsoap")
plt.xticks([i for i in range(1,13)])
plt.gca().axes.xaxis.set_ticklabels([])
plt.plot(data['month_number'],data['bathingsoap'],linewidth=3, color="black", marker = 'o', ms=8)

plt.subplot(2,1,2)
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("Sales data of a facewash")
plt.xticks([i for i in range(1,13)])
plt.plot(data['month_number'],data['facewash'],linewidth=3, color="red", marker = 'o', ms=8)

plt.show()
```
